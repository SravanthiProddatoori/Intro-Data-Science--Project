Numerical values are annotated within each cell, providing the actual correlation coefficient. Values closer to 1 or -1 indicate a stronger correlation, while values close to 0 suggest a weaker or no correlation.

# Machine Learning Model

# Logistic Regression

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix
from sklearn.preprocessing import LabelEncoder


# Data preprocessing
# Fill missing values if any
df.fillna(method='ffill', inplace=True)  # Filling missing values with forward fill

# Select features and target variable
selected_features = [
    'Route Type', 'Road Name', 'Collision Type', 'Weather', 'Surface Condition',
    'Light', 'Vehicle Movement', 'Vehicle Continuing Dir', 'Vehicle Going Dir',
    'Speed Limit', 'Vehicle Year', 'Vehicle Make', 'Vehicle Model',
    'Latitude', 'Longitude'
]

X = df[selected_features]  # Features
y = df['Injury Severity']  # Target variable

# Encode categorical variables using LabelEncoder
le = LabelEncoder()
X_encoded = X.copy()  # Create a copy to avoid modifying original data
for col in X.columns:
    if X[col].dtype == 'object':
        X_encoded[col] = le.fit_transform(X[col])

# Split data into train and test sets
X_train, X_test, y_train, y_test = train_test_split(X_encoded, y, test_size=0.2, random_state=42)

# Train the logistic regression model
model = LogisticRegression(max_iter=1000)  # Adjust max_iter as needed
model.fit(X_train, y_train)

# Predict on test set
y_pred = model.predict(X_test)

# Evaluate model performance
accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred, average='weighted')
recall = recall_score(y_test, y_pred, average='weighted')
f1 = f1_score(y_test, y_pred, average='weighted')
conf_matrix = confusion_matrix(y_test, y_pred)

print(f"Accuracy: {accuracy}")
print(f"Precision: {precision}")
print(f"Recall: {recall}")
print(f"F1 Score: {f1}")
print(f"Confusion Matrix:\n{conf_matrix}")
```

## Encoding

```python
le = LabelEncoder()
X_encoded = X.copy()  # Create a copy to avoid modifying original data
for col in X.columns:
    if X[col].dtype == 'object':
        X_encoded[col] = le.fit_transform(X[col])
```

## **MODEL BUILDING**

## **Splitting in train and test data**

```python
# Split data into train and test sets
X_train, X_test, y_train, y_test = train_test_split(X_encoded, y, test_size=0.2, random_state=42)
```

## Training and Prediction

```python
# Train the Random Forest model
model = RandomForestClassifier(n_estimators=100, random_state=42)  # You can adjust hyperparameters here
model.fit(X_train, y_train)

# Predict on test set
y_pred = model.predict(X_test)
```

#Overall design and the core features
The overall design of the dataset is to provide information on motor vehicle collisions that occurred on county and local roadways within Montgomery County. The data is collected through the Automated Crash Reporting System (ACRS) of the Maryland State Police and reported by various law enforcement agencies, including the Montgomery County Police, Gaithersburg Police, Rockville Police, or the Maryland-National Capital Park Police.

**Core Features of the Dataset:**

Motor Vehicle Collision Information: The dataset includes details of traffic collisions, providing insights into the circumstances, locations, and outcomes of these incidents.

Driver Information: The dataset focuses on motor vehicle operators (drivers) involved in the reported traffic collisions. It likely includes details about each driver, such as their identification, involvement in the collision, and potentially additional information about them.

Automated Crash Reporting System (ACRS): The dataset relies on the ACRS of the Maryland State Police for collecting crash-related data. This suggests a systematic and automated approach to reporting and recording collision information.

Reporting Parties and Law Enforcement Agencies: The dataset mentions reporting parties, indicating individuals or entities that provide information about the collisions. Additionally, it specifies the law enforcement agencies responsible for reporting and collecting the data, including the Montgomery County Police, Gaithersburg Police, Rockville Police, or the Maryland-National Capital Park Police.

Preliminary Information: The dataset explicitly states that the information provided is based on preliminary reports and may be subject to change after further investigation. This includes the possibility of unverified data, pending classifications, and the potential for updates based on additional information.

Data Quality Considerations: The dataset acknowledges the presence of information that is not yet verified, including the possibility of errors, both mechanical and human. This highlights the importance of understanding the limitations and potential inaccuracies in the data.

In summary, the dataset is designed to capture and report information on traffic collisions, with a focus on drivers involved in incidents on county and local roadways in Montgomery County. The core features include details about the collisions, driver information, the use of an automated reporting system, involvement of reporting parties, and the acknowledgment of preliminary and potentially evolving nature of the data.
